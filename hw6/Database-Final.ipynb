#### Part 4 For each pair of cities/airports determine how the daily change of high temperature and cloud cover from one city predicts the daily change of the other city 1,3, & 7 days in advance NOTE: Cloud cover from the Weather Underground tables ranges from 0 (clear) to 8 (completely cloudy)

```python
nports = 50
ndpoints = 1887 # hard code for now

## initialize two arrays
airportTemps = np.zeros((ndpoints,nports))
airportClouds = np.zeros((ndpoints,nports))

for col, airport in enumerate(db_info):
    ICAO = str(airport[0])

    # get non-weather data from other tables
    sql_cmd = ('SELECT ' + ICAO + '_WeatherData.max_temp, ' + 
    ICAO + '_WeatherData.cloud_cover from ' + ICAO + '_WeatherData')

    cursor.execute(sql_cmd)
    vectors = cursor.fetchall()
    vectors = np.asarray(vectors)
    temps = vectors[:,0]
    humidities = vectors[:,1] # the above three lines could be condensed but i don't care
    
    # store in array
    airportTemps[:,col] = temps[:1887]
    airportClouds[:,col] = humidities[:1887]
```

```python
# initialize arrays
TempCors = np.zeros((nports,nports,3)) # lags to be stacked in 3rd dim
CloudCors = np.zeros((nports,nports,3))
# list the lag in days
lags = [1,2,6] 

for a1 in range(nports):
    tempSeed = airportTemps[:,a1]
    CloudSeed = airportClouds[:,a1]
    
    for a2 in range(nports):
        tempTarg = airportTemps[:,a2]
        CloudTarg = airportClouds[:,a2]
        
        for page,lag in enumerate(lags):

            # lag the target airport by
            # the correct number of days
            LaggedTempTarg = tempTarg[lag:]
            LaggedCloudTarg = CloudTarg[lag:]
                
            # remove the correct number of 
            # days from the end of the 
            # seed timeseries
            LaggedTempSeed = tempSeed[:-lag]
            LaggedCloudSeed = CloudSeed[:-lag]
               
            # get corr and store in the big arrays
            tmp = np.corrcoef(LaggedTempTarg,LaggedTempSeed)
            TempCors[a1,a2,page] = tmp[0,1] # any off diagnoal
            tmp = np.corrcoef(LaggedCloudTarg,LaggedCloudSeed)                
            CloudCors[a1,a2,page] = tmp[0,1]
```

```python
def distance(point1,point2):
    # take two points, each (lat,lon)
    # and return euclidean distance
    point1 = np.asarray(point1)
    point2 = np.asarray(point2)
    diff = point1 - point2
    dist = np.sqrt(np.sum(diff**2))
    return dist
```

```python
# initialize distance matrix
airportDistances = np.zeros((nports,nports))
airportLongitudes = np.zeros((nports,nports))

for a1,airport1 in enumerate(db_info): 
    IC1 = str(airport1[0])
    for a2,airport2 in enumerate(db_info):
        IC2 = str(airport2[0])

        sql_cmd = ('SELECT latitude, longitude FROM TopAirportsData ' + 'WHERE ICAO = \'' + str(IC1) + '\'')
        cursor.execute(sql_cmd)
        point1 = np.asarray(cursor.fetchall())
        
        sql_cmd = ('SELECT latitude, longitude FROM TopAirportsData ' + 'WHERE ICAO = \'' + str(IC2) + '\'')
        cursor.execute(sql_cmd)
        point2 = np.asarray(cursor.fetchall())        
        
        airportDistances[a1,a2] = distance(point1,point2)
#         airportLongitudes[a1,a2] = point1[0][1]-point2[0][1]
        airportLongitudes[a1,a2] = np.abs(point1[0][1] - point2[0][1])
```

```python
#thanks to DAN BLISS
def top_ten(matrix):
    """Return indices for the ten elements of matrix, not along the diagonal, with the largest values.
    
    Parameters
    ----------
    matrix : numpy array
        Two-dimensional array.
    
    Returns
    -------
    indices : list of tuples
        List of (row, column) indicies of the ten largest values in matrix, in order from largest to smallest.
        
    values : list of floats
        The largest elements in matrix, in order from largest to smallest.

    """
    x, y = np.unravel_index(np.argsort(matrix, axis=None), matrix.shape)
    indices = zip(x, y)[::-1]
    indices = [(x, y) for x, y in indices if x != y][:10]
    values = [matrix[x, y] for x, y in indices]
    return indices, values
```
