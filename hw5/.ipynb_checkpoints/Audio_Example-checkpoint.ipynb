##### Cool! That worked. However, after some advice from Lauren Weiss, this is not a great way to detect spikes and choose notes one by one. Let's take another look at the power spectrum but not by using the spectrogram. We have the fourier transform in the fftData variable.

```python
# take the fft again
# (:40K is so we don't get the transient from when the wave turns off)
fftData = np.fft.rfft(the_wave) # rfft is just the real part (40000)
fftData = abs(fftData**2)
# humans hear 20Hz to 20kHz but piano notes 
# are usually in between 20 and 5000Hz
plt.plot(fftData[20:5000])
```

Better.  Let's zoom to just 20-1500

```python
plt.plot(fftData[1000:4000])
```

##### Now we can see the peak at around 400Hz (after translation to Hz by idx*rate/chunk) and the other spikes are unlikely to be other notes. I think what we want to do now is look for the highest spike, classify it as a note, and then move to the next lowest spike and repeat until we hit some sort of floor. In this example, a noise floor of 1x10e25 would work. However, we need a more flexible criterion.  So let's try taking the max power, and then going down the spikes until you get to 1/4th the power of the maximum spike. Also, let's load up one of the files with more than one note.

```python
filename = path + 'C4+A4_PopOrgan.aif'
file = aifc.open(filename,'r')

# get some info about the file
# total number of samples 
nframes = file.getnframes()

# sample width
swidth = file.getsampwidth()

# number of channels 
nchannels = file.getnchannels()

# samplign frequency is made 
# available by a function
fs = file.getframerate()

# total number of samples 
frames = file.readframes(nframes)

# actually read the wave (found int32 is the right 
# code for getting the right length wave out)
the_wave = np.fromstring(frames,np.int32).byteswap()
time = range(len(the_wave))

the_wave = the_wave[:len(the_wave)/2]
rate = fs
chunk = len(the_wave) #just try one big window to start

window = np.blackman(chunk) 

# multiply data by the window
indata = the_wave*window

# Take the fft and square each value
fftData=abs(np.fft.rfft(indata))**2
plt.plot(fftData[:5000])
axhline(y= max(fftData[:5000])/2,linewidth = 4,color = 'r')
```

##### Function to just return note name given a frequency

```python
def find_note(freq):
    diffs = np.asarray(noteFqs) - freq
    diffs = abs(diffs)
    idx = diffs.argmin()
    note = notes[idx-1]
    return note
```

```python
# preprocessing
window = np.blackman(len(the_wave))

# multiply data by the window
indata = the_wave*window

# Take the fft and square each value
fftData=abs(np.fft.rfft(indata))**2
    
# just look under 5000
fftData = fftData[:5000]
```

```python
# current max
max_spike = max(fftData) # let's assume all the notes we're getting are on the piano
noise_floor = max_spike/2
above_floor = 1
```

```python
import sys
while above_floor:
    # identify note at first max

    # find the maximum
    which = fftData.argmax()

    # find the frequency and output it
    thefreq = (which)*rate/chunk
    note = find_note(thefreq)
    print note + ' detected in wave (' + str(thefreq) + 'Hz) with power ' + str(fftData[which])
    
    # remove spike
    fftData[(which-200):(which+200)] = 0
    #fftData[:(which+200)] = 0
    
    # find new highest spike
    # print 'current max is ' + str(max(fftData))
    if  max(fftData) < noise_floor:
        print 'note analysis complete'
        above_floor = 0
```
